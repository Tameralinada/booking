 # Null Values ( Drop - fill )
 ## agent (float ) 16340 --- > fill ( mean - mode - median )

## country ( Object ) 488 ---> drop (missing values ) \\ fill ( mode )

```python
data_cleaned = data1().dropna()
```

```python
data_cleaned
```

```python
data1 = data1.drop('company' , axis = 1)
```

```python
data1
```

```python
#data1['agent'] = data1['agent'].fillna(0)
data1['agent'] = data1['agent'].fillna(data['agent'].mean())
```

```python
data1.isnull().sum()
```

```python
data1['country'] = data1['country'].fillna(data1['country'].mode()[0])
```
